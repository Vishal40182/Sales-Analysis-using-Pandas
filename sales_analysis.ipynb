```python
path = r'./sales analysis/Pandas-Data-Science-Tasks-master/SalesAnalysis/Sales_Data'
all_files = glob.glob(path + "/*.csv")

df1 =[]
for file in all_files:
    df = pd.read_csv(file, index_col=None, header=0)
    df1.append(df)
    
total_data = pd.concat(df1, axis=0, ignore_index=True)
```

```python
total_data.info()
```

```python
total_data.head()
```

```python
total_data["month"] = total_data["Order Date"].str[0:2]
total_data.head()
```

```python
null_row = total_data[total_data.isna().any(axis = 1)]
null_row.head()
```

```python
total_data = total_data.dropna(how = "all")
total_data.shape
```

```python
temp = total_data[total_data["month"] == "Or"]
temp.shape[0]
```

```python
total_data = total_data[total_data["month"] != "Or"]
total_data.shape[0]
```

```python
total_data["month"] = total_data["Order Date"].str[0:2]
total_data["month"] = total_data["month"].astype('int32')
total_data.tail()
```

## Best month for sales 

```python
total_data.info()
```

```python
total_data["Quantity Ordered"] = total_data["Quantity Ordered"].astype('int32')
total_data["Price Each"] = total_data["Price Each"].astype('float64')
total_data["total_sales"] = (total_data["Quantity Ordered"]*total_data["Price Each"])
total_data.head()
```

```python
best_month = total_data.groupby("month")
best_month = best_month["total_sales"].sum()
best_month
```

### Note: December is the best month for sales.
